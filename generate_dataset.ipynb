```python
!pip install faker

from faker import Faker
import pandas as pd
import random
from collections import Counter

fake = Faker()

def gerar_linha():
    temperatura = round(random.uniform(-10, 45), 2)
    umidade = round(random.uniform(10, 100), 2)
    pressao = round(random.uniform(950, 1050), 2)
    vento = round(random.uniform(0, 25), 2)
    chuva = round(random.uniform(0, 300), 2)
    nivel_agua = round(random.uniform(0, 5), 2)
    gases = round(random.uniform(300, 2000), 2)
    luminosidade = round(random.uniform(0, 100000), 2)

    if chuva > 100 and nivel_agua > 2:
        evento = "enchente"
    elif umidade < 20 and temperatura > 35:
        evento = "incendio"
    elif pressao < 980 and vento > 10:
        evento = "tempestade"
    elif temperatura > 38:
        evento = "onda_calor"
    else:
        evento = "normal"

    return {
        "temperatura": temperatura,
        "umidade": umidade,
        "pressao": pressao,
        "vento": vento,
        "chuva": chuva,
        "nivel_agua": nivel_agua,
        "gases": gases,
        "luminosidade": luminosidade,
        "evento": evento
    }

# Gerar dataset original
dataset = [gerar_linha() for _ in range(1000)]

count_events = Counter(row['evento'] for row in dataset)
print("Antes do balanceamento:", count_events)

max_count = max(count_events.values())

# Criar lista por classe
groups = {}
for row in dataset:
    groups.setdefault(row['evento'], []).append(row)

# Balancear
balanced_dataset = []
for evento, items in groups.items():
    needed = max_count - len(items)
    balanced_dataset.extend(items)
    for _ in range(needed):
        balanced_dataset.append(gerar_linha())

count_events_after = Counter(row['evento'] for row in balanced_dataset)
print("Depois do balanceamento:", count_events_after)

df = pd.DataFrame(balanced_dataset)

df.to_csv("dataset_desastres_faker.csv", index=False)

print("CSV salvo como 'dataset_desastres_faker.csv'")
```
